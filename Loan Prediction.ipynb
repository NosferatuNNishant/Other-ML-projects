#### Columns that need value imputation

```python
## Gender, Married, Dependents, Self_Employed, LoanAmount, Loan_Amount_Term, Credit_History 
cat_v = ['Gender', 'Married', 'Self_Employed']
num_v = ['LoanAmount', 'Loan_Amount_Term', 'Credit_History']
```

#### Missing value imputation

```python
def mvimputation(df):
    for col in cat_v:
        df[col] = df[col].fillna('mode')
    for col in num_v:
        df[col] = df[col].fillna(df[col].mean())
```

```python
mvimputation(df)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df1 = df.select_dtypes('object') ## made a subset of the categorical variables to label encode separately
```

```python
df1.head(2)
```

```python
df1 = df1.drop('Loan_ID', axis =1 )
```

```python
df1.head(2)
```

```python
df1.columns
```

```python
cols_df1 = ['Gender', 'Married', 'Dependents', 'Education', 'Self_Employed', 'Property_Area', 'Loan_Status']

for col in cols_df1:
    df1[col] = le.fit_transform(df1[col])    
```

```python
cols_not_in_df1 = [col for col in df.columns if col not in df1.columns]
print(cols_not_in_df1)
```

```python
df2 = df[['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term', 'Credit_History']]    ## remove loan id
```

```python
df2.head()
```

```python
DS = pd.concat([df1, df2], axis = 1)
```

```python
DS.head()   ## this is my final dataset using which I will make the prediction
```

```python
X = DS.drop('Loan_Status', axis = 1)
y = DS['Loan_Status']
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.15)
```

```python
X_train.head(3)
```

## Logistic Regression

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train , y_train)
```

```python
y_train_pred_log = lr.predict(X_train)
print(accuracy_score(y_train, y_train_pred_log))  ### 81% accuracy on train dataset using Logistic regression
```

```python
y_test_pred_log = lr.predict(X_test)
print(accuracy_score(y_test_pred_log, y_test))  ## 79% accuracy on test dataset
```
