```python
cols1 = ['index', 'budget', 'genres', 'homepage', 'id', 'keywords',
       'original_language', 'original_title', 'overview', 'popularity',
       'production_companies', 'production_countries', 'release_date',
       'revenue', 'runtime', 'spoken_languages', 'status', 'tagline', 'title',
       'vote_average', 'vote_count', 'cast', 'crew', 'director']
cols2 = ['budget', 'genres','keywords', 'original_language','popularity', 'production_companies','production_countries',
        'revenue','tagline', 'vote_average','vote_count', 'cast', 'director']
```

```python
print(cols2 in cols1)
```

##### I am using cols2 as the columns for my recommendation system

```python
for col in cols2:
    df[col] = df[col].fillna('')        ### fill the empty strings
```

### Combining the features

```python
len(cols2)
```

```python
comb_feat = ''
for i in cols2:
    comb_feat += df[i].astype(str) + '_'
```

```python
comb_feat
```

```python
vec = TfidfVectorizer()
```

```python
vec_feat = vec.fit_transform(comb_feat)      ## vectorized features
```

```python
print(vec_feat)
```

```python
similarity = cosine_similarity(vec_feat)
```

```python
print(similarity)
```

## Getting Input from the User

```python
movie_name = input('For the movie for which you want to find similar movies, enter its name:')
```

#### List of all movie names

```python
movies_list  = df['title'].tolist()
```

```python
movies_list
```

### Finding close match

```python
find_close_match = difflib.get_close_matches(movie_name, movies_list)
print(find_close_match)          ### these 3 movies are close to Iron man
```

```python
mov_name = 'Man of Steel'
find_close_match = difflib.get_close_matches(mov_name, movies_list)
print(find_close_match)          ### not getting good results here
```

#### Getting a similarity score

```python
close_match = find_close_match[0]
print(close_match)
```

```python
# finding the index of the movie with title

index_of_the_movie = df[df.title == close_match]['index'].values[0]
print(index_of_the_movie)
```

```python
similarity_score = list(enumerate(similarity[index_of_the_movie]))
print(similarity_score)
```
